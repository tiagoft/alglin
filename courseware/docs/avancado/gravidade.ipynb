# Estimando a Gravidade

Aprendemos a fazer um processo chamado *regressão*. Uma das aplicações desse processo é encontrar (ou estimar) valores para grandezas físicas à partir de observações. Nesta atividade, vamos estimar a *aceleração da gravidade*.

## Teoria

*Teoria* é a maneira que acreditamos que algum fenômeno se comporta. Por exemplo: a Teoria da Gravidade é a ideia de que corpos tendem a se atrair seguindo algumas equações específicas. No contexto de um experimento, *teoria* se refere aos modelos matemáticos que regem os fenômenos que esperamos encontrar, e entendê-los implica em uma série de decisões práticas sobre os experimentos.

Existem muitas maneiras de estimar a aceleração da gravidade. Vamos estudar uma delas.

### Movimento Uniformemente Variado

Nas nossas aulas de física do Ensino Médio, provavelmente já tivemos contato com a equação do Movimento Uniformemente Acelerado:

$$
s = s_0 + v_0 t + \frac{a t^2}{2}.
$$

Podemos usar essa equação para calcular a posição de um objeto em queda livre (partindo do repouso e ignorando a resistência do ar) fazendo $s_0=0$, $v_0=0$ e $a=g$, ficando com:

$$
s = \frac{a t^2}{2}.
$$

### Regressão Linear com pseudo-inversa

Se temos algum fenômeno do tipo $y=ax$, e podemos medir várias entradas $x$ e saídas correspondentes $y$, então temos:

$$
\begin{bmatrix} \hat{y}_1 \\ \hat{y}_2 \\ \vdots \\ \hat{y}_n \end{bmatrix} = \begin{bmatrix} x_1 \\ x_2 \\ \vdots  \\ x_n  \end{bmatrix}  a 
$$

ou:

$$
\hat{y} = x a
$$

Podemos então estimar $a$ usando a pseudo-inversa:
$$
\begin{array}{rl}
x^T y &= (x^T x) a \\
(x^T x)^{-1} x^T y &= (x^T x)^{-1} (x^T x) a \\
(x^T x)^{-1} x^T y &= a \\
\end{array}
$$

### Estimação do modelo de queda livre

O uso de regressão linear implica em um fenômeno $y=ax$, mas a equação da queda livre modela um termo quadrático $s=\frac{at^2}{2}$. 

Porém, podemos usar uma variável auxiliar $\tau = \frac{t^2}{2}$, de forma que:

$$
s=a \tau
$$

Portanto, podemos estimar $a$ através da formulação matricial:

$$
\begin{bmatrix} s_1 \\ s_2 \\ \vdots \\ s_n \end{bmatrix} = \begin{bmatrix} \tau_1 \\ \tau_2 \\ \vdots  \\ \tau_n  \end{bmatrix}  a
$$

## Método

Veja como a *teoria* implica imediatamente em um método para encontrar a aceleração da gravidade:

1. Para usar a equação matricial, precisamos de vários valores de $s_i$ e $t_i$ correpondentes. Portanto, precisamos *medir o tempo de queda para diferentes alturas e anotar o resultado*.
2. Para usar a formulação linear, precisamos calcular $\tau_i = t^2_i/2$ para cada valor de tempo medido
3. Após, usamos a pseudo-inversa para encontrar o valor de $a$.

!!! info "Antes de prosseguir..."
    O método para encontrar $a$ só é eficaz se você for capaz de responder *por que* cada uma das etapas é relevante. Discuta com sua equipe *por que* cada uma das etapas do método é importante!
