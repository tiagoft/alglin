## Exercício 2
**Objetivo: manipular probabilidades relacionadas a Cadeias de Markov**

Vamos agora extrapolar a situação anterior para o caso de termos $n$ estações.

Vamos chamar de $P(e_t = j)$ a probabilidade de uma bicicleta estar na estação $j$ após $t$ iterações do nosso sistema.

Seguindo essa mesma notação, temos que $P(e_t = j | e_{t-1} = i)$ é a probabilidade de uma bicicleta estar na estação $j$ na iteração $t$ *dado que sabemos que* ela estava na estação $i$ na iteração $t-1$.

Como estar em um determinado estado (ou: uma bicicleta estar em uma estação) leva necessariamente a não estar em outro estado (isto é: uma bicicleta só pode estar em uma estação de cada vez), então temos que:

$$
P(e_t = j) = \sum_{i} P(e_t = j | e_{t-1} = i) P(e_{t-1} = i)
$$

(a notação $\sum_i$ representa "somatória para todo $i$ válido").

(a) Demonstre que $P(e_t = j) = \sum_{i} P(e_t = j | e_{t-1} = i) P(e_{t-1} = i)$ (dica: lembre-se da definição de probabilidade condicional e desenhe um diagrama de Venn desse caso).

(b) Demonstre que é possível calcular $P(e_t = 1)$ usando a forma matricial:

$$
\begin{bmatrix} P(e_t = 1) \end{bmatrix} = \begin{bmatrix} P(e_t = 1 | e_{t-1} = 1) & P(e_t = 1 | e_{t-1} = 2) & \cdots & P(e_t = 1 | e_{t-1} = n) \end{bmatrix} \begin{bmatrix} P(e_{t-1} = 1) \\ P(e_{t-1} = 2) \\ \vdots \\ P(e_{t-1} = n) \end{bmatrix}
$$

(c) Demonstre que podemos calcular $P(e_t = j)$ para todo $j$ usando a equação:

$$
x_{t} = A x_{t-1},
$$
onde:

$$
x_{t} = \begin{bmatrix} P(e_{t} = 1) \\ P(e_{t} = 2) \\ \vdots \\ P(e_{t} = n) \end{bmatrix},
$$

$$
x_{t-1} = \begin{bmatrix} P(e_{t-1} = 1) \\ P(e_{t-1} = 2) \\ \vdots \\ P(e_{t-1} = n) \end{bmatrix},
$$

e

$$
A = \begin{bmatrix} a_{i,j} \end{bmatrix}, a_{i,j} = P(e_t = j | e_{t-1} = i)
$$

(d) Se $A$ é uma matriz que leva as probabilidades de estar em cada estação numa iteração para essas mesmas probabilidades na iteração seguintes, que matriz é capaz de calcular as probabilidades de estar em estações duas iterações adiante?

**Objetivo: calcular auto-valores e auto-vetores usando pacotes Python**

O problema das bicicletas de Montreal está ligado a uma característica de matrizes que é a existência de vetores e valores chamados de auto-valores e de auto-vetores. Quando um auto-vetor de uma matriz é multiplicado pela própria matriz, o resultado é um múltiplo do próprio vetor, isto é:

$$
Ax = x \lambda,
$$
onde: $A$ é a matriz, $x$ é o auto-vetor e $\lambda$ é o auto-valor correspondente.

Por exemplo:

$$
\begin{bmatrix} 
    2 & 0  \\
    0 & 1 
\end{bmatrix}
\begin{bmatrix} 
    1 \\
    0  
\end{bmatrix}
= 
\begin{bmatrix} 
    2 \\
    0  
\end{bmatrix}
=
2
\begin{bmatrix} 
    1 \\
    0  
\end{bmatrix},
$$
portanto $\begin{bmatrix}1\\0\end{bmatrix}$ é um auto-vetor de $\begin{bmatrix} 2 & 0 \\  0 & 1 \end{bmatrix}$, e $2$ é o auto-valor associado.

Se tivermos algum vetor na mesma direção de $x$ mas multiplicado por algum número real $\alpha$, isto é, $y=\alpha x$, podemos manter a mesma equação:

$$
A \alpha x = \alpha x \lambda \Rightarrow Ay = y \lambda.
$$

Se você quiser calcular auto-valores e auto-vetores manualmente, pode usar. Porém, essa conta é muito longa e por isso usaremos um pacote computacional para isso.

**Objetivo: auto-vetores e auto-valores como pontos de estabilidade, expansão, ou colapso**

 Vamos retomar o problema da população de carcarás e de sapos. Nesse problema, tínhamos a situação em que carcarás estão no mesmo habitat que sapos, obedecendo às seguintes regras:

* A população de carcarás naturalmente cai em 20% a cada mês
* Para cada cinco mil sapos existentes naquele mês, um novo carcará nasce
* A população de sapos naturalmente sobe em 10% porque eles se reproduzem
* A cada mês, cada carcará consegue comer 100 sapos

Isso significa que podemos escrever equações para calcular o número de carcarás e de sapos a cada mês, tomando por base a quantidade deles no mês anterior:

$$ 
\begin{cases}
    \begin{aligned}
    c_t & = 0.8 c_{t-1} + 0.2 s_{t-1} \\
    s_t & = - 0.1 c_{t-1} + 1.1 s_{t-1} \\
    \end{aligned}
\end{cases}
$$

Podemos fazer esse cálculo usando a operação matricial:
$$
\begin{bmatrix}
    c_t \\
    s_t 
\end{bmatrix}
=
\begin{bmatrix}
    0.8 & 0.2\\
    -0.1 & 1.1
\end{bmatrix} 
\begin{bmatrix}
    c_{t-1} \\
    s_{t-1}
\end{bmatrix}
$$

a) Calcule os auto-vetores e auto-vetores da matriz $A = \begin{bmatrix}
    0.8 & 0.2\\
    -0.1 & 1.1
\end{bmatrix}$.

b) Iniciando com uma população aleatória de carcarás e sapos, calcule o vetor $x_{t}$ com a população de carcarás e sapos após muitos meses da iteração nessa comunidade.

c) Encontre o ângulo entre $x_t$ e cada um dos auto-vetores de $A$. Qual é o auto-vetor de $A$ mais alinhado a $x_t$? Qual é o auto-valor associado a ele?

```python
import matplotlib.pyplot as plt
import numpy as np

A = np.array([[0.8, 0.2], [-0.1,  1.1]])
x = np.array([[600], [500]])
populacao_carcaras = []
populacao_sapos = []
for _ in range(1000):
    populacao_carcaras.append(x[0,0])
    populacao_sapos.append(x[1,0])
    x = A @ x

plt.figure(figsize=(7,2))
plt.plot(populacao_carcaras, label='Carcarás')
plt.plot(populacao_sapos, label='Sapos (* 1000)')
plt.ylim([0,800])
plt.ylabel('População')
plt.xlabel('Semanas')
plt.show()
```
