3) A coxinha  5 reais e o brigadeiro  2 reais

```python
c = np.linspace(0,10,3)
b = 17 - 3*c

b2 = 12 - 2*c

A = np.array([[3,1], [2,1]])
y = np.array([[17], [12]])

print(np.linalg.solve(A,y))
# coxinha  5 reais e o brigadeiro  2 reais


plt.figure()
plt.plot(c, b) 
plt.plot(c, b2) 
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
